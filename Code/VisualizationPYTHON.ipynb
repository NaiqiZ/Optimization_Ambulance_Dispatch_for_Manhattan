```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 读取数据
df = pd.read_csv("cost_matrix_sijWaitij.csv", index_col=0)

# 创建一个掩码，仅显示非零值
mask = df != 0

# 绘制热力图
plt.figure(figsize=(20, 10))  # 设置更宽的图像比例
sns.heatmap(
    df,
    annot=df.where(mask),  # 仅标注非零值
    fmt="g",
    cmap="YlGnBu",  # 柔和的配色方案
    cbar_kws={'label': 'Cost (Time)'},
    mask=~mask,  # 隐藏零值
    linewidths=0.5,  # 设置网格线宽度
    linecolor="gray",  # 设置网格线颜色
    annot_kws={"size": 10},  # 调整标注文字大小
)

# 设置标题和轴标签
plt.title("Improved Cost Matrix Heatmap (Non-zero values only)", fontsize=20, pad=20)
plt.xlabel("Zipcodes", fontsize=16)
plt.ylabel("Ambulance Areas", fontsize=16)
plt.xticks(fontsize=12, rotation=45)  # 调整横轴刻度文字大小和角度
plt.yticks(fontsize=12)

# 保存图像
plt.savefig("improved_cost_matrix_heatmap.png", dpi=300, bbox_inches="tight")

# 显示图像
plt.show()
```
