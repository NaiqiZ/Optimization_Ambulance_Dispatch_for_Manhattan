```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# è¯»å–æ•°æ®
df = pd.read_csv("cost_matrix_sijWaitij.csv", index_col=0)

# åˆ›å»ºä¸€ä¸ªæŽ©ç ï¼Œä»…æ˜¾ç¤ºéžé›¶å€¼
mask = df != 0

# ç»˜åˆ¶çƒ­åŠ›å›¾
plt.figure(figsize=(20, 10))  # è®¾ç½®æ›´å®½çš„å›¾åƒæ¯”ä¾‹
sns.heatmap(
    df,
    annot=df.where(mask),  # ä»…æ ‡æ³¨éžé›¶å€¼
    fmt="g",
    cmap="YlGnBu",  # æŸ”å’Œçš„é…è‰²æ–¹æ¡ˆ
    cbar_kws={'label': 'Cost (Time)'},
    mask=~mask,  # éšè—é›¶å€¼
    linewidths=0.5,  # è®¾ç½®ç½‘æ ¼çº¿å®½åº¦
    linecolor="gray",  # è®¾ç½®ç½‘æ ¼çº¿é¢œè‰²
    annot_kws={"size": 10},  # è°ƒæ•´æ ‡æ³¨æ–‡å­—å¤§å°
)

# è®¾ç½®æ ‡é¢˜å’Œè½´æ ‡ç­¾
plt.title("Improved Cost Matrix Heatmap (Non-zero values only)", fontsize=20, pad=20)
plt.xlabel("Zipcodes", fontsize=16)
plt.ylabel("Ambulance Areas", fontsize=16)
plt.xticks(fontsize=12, rotation=45)  # è°ƒæ•´æ¨ªè½´åˆ»åº¦æ–‡å­—å¤§å°å’Œè§’åº¦
plt.yticks(fontsize=12)

# ä¿å­˜å›¾åƒ
plt.savefig("improved_cost_matrix_heatmap.png", dpi=300, bbox_inches="tight")

# æ˜¾ç¤ºå›¾åƒ
plt.show()
```
