```python
# Create KDE for each list
kde_minmax = gaussian_kde(MinMax_everyincident_waittime)
kde_minsum = gaussian_kde(MinSum_everyincident_waittime)

# Define a range of values for the x-axis
x_min = min(min(MinMax_everyincident_waittime), min(MinSum_everyincident_waittime))
x_max = max(max(MinMax_everyincident_waittime), max(MinSum_everyincident_waittime))
x_range = np.linspace(x_min, x_max, 500)

# Evaluate the KDE on the x-axis range
y_minmax = kde_minmax(x_range)
y_minsum = kde_minsum(x_range)

# Plot the KDE curves
plt.figure(figsize=(10, 6))
plt.plot(x_range, y_minmax, label='Min-Max Model', linewidth=2)
plt.plot(x_range, y_minsum, label='Min-Sum Model', linewidth=2)

# Add title and labels
plt.title('KDE Distribution of Wait Times(Robust Models)', fontsize=16)
plt.xlabel('Wait Time', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend(fontsize=20)

# Add grid for better visualization
plt.grid(linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
